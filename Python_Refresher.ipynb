```python
# Question 1
# Used the variable days to store the input from the user.
Days = int(input("Enter the number of days: ")) # a string displays along side the input box
seconds = Days * 24 * 60 * 60 # this will calculate the seconds in a day based on the input from the user
print("There are " + str(seconds) + " seconds in " + str(Days) + " days.")
```

```python

# Question 2
# This will ask the user for the radius
radius = float(input("Enter Radius: "))
volume = (4/3) * 3.14 * (radius ** 3) # this will calculate the volume based on the value input of radius

print(f"The volume of a sphere is {volume}")
```

```python
# Question 3
def area_of_square(side): #function for the area with a parameter called side
    return side ** 2

def perimeter_of_square(side): #perimeter function with a parameter called side
    return 4 * side

side = float(input("Side length of the square: ")) # user input is stored in the variable "side"
area = area_of_square(side) # the value of side is passed into the previously defined functions as an argument
perimeter = perimeter_of_square(side)
#final output
print(f"Square side length is {side}:")
print("SO")
print(f"Area = {area}")
print(f"Perimeter = {perimeter}")
```

```python
# Question 4
# Function to check case
def check_case(ch):
    if ch.isupper():
        return "Uppercase"
    elif ch.islower():
        return "Lowercase"
    else:
        return "Not a letter"

char = input("Enter a single character: ")

result = check_case(char)

print(f"The character '{char}' is {result}.")
```

```python
# Question 5
x = 0 #variables x and y are set/initialized to the given values
y = 20

# Used a while loop that will continue until y is less than or equal to 6
while y >= 6:
    y = y - 4          # subtract 4 from y
    x = x + (2 / y)    # add 2/y to x

# Final value of x
print(f"Final value of x is {x}")
```

```python
# Question 6

numbers = [] #initialize array


for i in range(5): # will prompt the user to input 5 values only.
    num = float(input(f"Enter value {i+1}: "))
    numbers.append(num) # adds the values to the array.

# will calculate the average of the 5 values
average = sum(numbers) / len(numbers)
print("For the following:", numbers)
print(f"Average of the values is {average:.2f}")
```
